```python
import os
import json
from sentence_transformers import SentenceTransformer
import numpy as np

def split_text_into_chunks_by_char(text, chunk_size=2000):
    """Split the text into chunks based on character count."""
    chunks = []
    # Split the text into chunks of the given character size
    for i in range(0, len(text), chunk_size):
        chunks.append(text[i:i + chunk_size])

    return chunks

def process_json_file(file_path):
    """Process the JSON file: Extract keys and chunk values."""
    with open(file_path, 'r') as file:
        data = json.load(file)

    # Extract keys (section titles)
    keys = list(data.keys())

    # Prepare the chunked values
    chunked_values = {}
    for key, value in data.items():
        chunks = split_text_into_chunks_by_char(value)
        chunked_values[key] = chunks

    return keys, chunked_values

def embed_chunks(chunked_values):
    """Embed the chunks using SentenceTransformer model."""
    # Use a more advanced embedding model if needed
    model = SentenceTransformer('all-MiniLM-L6-v2')

    embeddings = {}

    for key, chunks in chunked_values.items():
        # Embed each chunk
        chunk_embeddings = model.encode(chunks)
        embeddings[key] = chunk_embeddings.tolist()  # Convert to list for JSON serialization

    return embeddings

def save_to_json(keys, chunked_values, embeddings, output_dir, filename):
    """Save the keys, chunked values, and embeddings into separate JSON files."""
    # Save keys (section titles)
    with open(os.path.join(output_dir, f"{filename}_keys.json"), 'w') as key_file:
        json.dump(keys, key_file, indent=4)

    # Save chunked values
    with open(os.path.join(output_dir, f"{filename}_chunked_values.json"), 'w') as chunk_file:
        json.dump(chunked_values, chunk_file, indent=4)

    # Save embeddings
    with open(os.path.join(output_dir, f"{filename}_embeddings.json"), 'w') as embed_file:
        json.dump(embeddings, embed_file, indent=4)


def process_all_json_files(directory_path, output_dir):
    """Process all JSON files in the directory and split them."""
    for filename in os.listdir(directory_path):
        if filename.endswith(".json"):
            file_path = os.path.join(directory_path, filename)
            print(f"Processing file: {filename}")

            keys, chunked_values = process_json_file(file_path)

            # Embed the chunks
            embeddings = embed_chunks(chunked_values)

            # Save to JSON files
            save_to_json(keys, chunked_values, embeddings, output_dir, filename.split('.')[0])

            # Print embeddings details
            print("\n--- Embedding Details ---")
            for key, chunks_embeddings in embeddings.items():
                print(f"\nKey: {key}")
                print(f"Number of chunks: {len(chunks_embeddings)}")

                # Print details of first few embeddings
                for i, embedding in enumerate(chunks_embeddings[:3]):
                    print(f"\nChunk {i+1} Embedding:")
                    print(f"  Dimensions: {len(embedding)}")
                    print(f"  First 5 values: {embedding[:5]}")
                    print(f"  Norm (magnitude): {np.linalg.norm(embedding):.4f}")

                print("-" * 50)

# Path to the folder containing JSON files
directory_path = "/content/drive/MyDrive/Colab Notebooks/LLM Hackathon/microlabs_usa"  # Replace with your directory
output_dir = "/content/drive/MyDrive/Colab Notebooks/LLM Hackathon/chunked_microlabs_usa"  # Replace with your output directory

# Ensure output directory exists
os.makedirs(output_dir, exist_ok=True)

# Process all JSON files
process_all_json_files(directory_path, output_dir)
print("Processing complete!")
```

```python
# Install required libraries
!pip install -q chromadb sentence-transformers openai

import os
import json
import chromadb
from sentence_transformers import SentenceTransformer
import numpy as np

class ChromaEmbeddingManager:
    def __init__(self, embedding_model='all-MiniLM-L6-v2'):
        """
        Initialize Chroma DB embedding manager

        :param embedding_model: Sentence Transformer model to use
        """
        # Initialize embedding model
        self.embedding_model = SentenceTransformer(embedding_model)

        # Initialize Chroma client (using persistent storage in Google Drive)
        self.chroma_client = chromadb.PersistentClient(path="/content/drive/MyDrive/chroma_db")

    def create_collection(self, collection_name='microlabs_usa'):
        """
        Create or get a Chroma collection

        :param collection_name: Name of the collection
        :return: Chroma collection
        """
        return self.chroma_client.get_or_create_collection(name=collection_name)

    def load_processed_embeddings(self, embeddings_dir):
        """
        Load processed embeddings from JSON files

        :param embeddings_dir: Directory containing processed embedding files
        :return: Dictionary of processed embeddings and their metadata
        """
        # Processed embedding files
        keys_file = [f for f in os.listdir(embeddings_dir) if f.endswith('_keys.json')][0]
        chunks_file = [f for f in os.listdir(embeddings_dir) if f.endswith('_chunked_values.json')][0]
        embeddings_file = [f for f in os.listdir(embeddings_dir) if f.endswith('_embeddings.json')][0]

        # Full file paths
        keys_path = os.path.join(embeddings_dir, keys_file)
        chunks_path = os.path.join(embeddings_dir, chunks_file)
        embeddings_path = os.path.join(embeddings_dir, embeddings_file)

        # Load files
        with open(keys_path, 'r') as f:
            keys = json.load(f)
        with open(chunks_path, 'r') as f:
            chunked_values = json.load(f)
        with open(embeddings_path, 'r') as f:
            embeddings = json.load(f)

        return keys, chunked_values, embeddings

    def populate_chroma_db(self, embeddings_dir):
        """
        Populate Chroma DB with embeddings

        :param embeddings_dir: Directory with processed embedding files
        :return: Populated Chroma collection
        """
        # Load processed embeddings
        keys, chunked_values, embeddings = self.load_processed_embeddings(embeddings_dir)

        # Create collection
        collection = self.create_collection()

        # Add embeddings to Chroma DB
        for key, chunks_embeddings in embeddings.items():
            for idx, (chunk, embedding) in enumerate(zip(
                chunked_values[key],
                chunks_embeddings
            )):
                unique_id = f"{key}_chunk_{idx}"
                collection.add(
                    ids=[unique_id],
                    embeddings=[embedding],
                    metadatas=[{
                        "key": key,
                        "chunk_index": idx,
                        "text": chunk
                    }]
                )

        print(f"Added {collection.count()} embeddings to Chroma DB")
        return collection

    def semantic_search(self, collection, query, top_k=3):
        """
        Perform semantic search on the collection

        :param collection: Chroma DB collection
        :param query: Search query
        :param top_k: Number of top results to retrieve
        :return: Retrieved documents
        """
        # Generate query embedding
        query_embedding = self.embedding_model.encode([query])[0].tolist()

        # Perform similarity search
        results = collection.query(
            query_embeddings=[query_embedding],
            n_results=top_k
        )

        return results['metadatas'][0]

def main():
    # Set paths (adjust to your Google Drive structure)
    embeddings_dir = "/content/drive/MyDrive/Colab Notebooks/LLM Hackathon/chunked_microlabs_usa"

    # Initialize Chroma Embedding Manager
    em = ChromaEmbeddingManager()

    # Populate Chroma DB
    collection = em.populate_chroma_db(embeddings_dir)

    # Example queries
    queries = [
        "What is the composition and primary use of Paracetamol?",
        "Can I take Ibuprofen if I have a history of stomach ulcers?",

    ]

    # Perform semantic search for each query
    for query in queries:
        print("\nQuery:", query)
        results = em.semantic_search(collection, query)

        # Print retrieved documents
        for idx, doc in enumerate(results, 1):
            print(f"\nRetrieved Document {idx}:")
            print(f"Key: {doc['key']}")
            print(f"Chunk Index: {doc['chunk_index']}")
            print("Text Snippet:", doc['text'][:300] + "..." if len(doc['text']) > 300 else doc['text'])

if __name__ == '__main__':
    main()
```

```python
# Install necessary libraries
!pip install streamlit sentence-transformers transformers hdbscan

import os
import json
import pandas as pd
from sklearn.neighbors import NearestNeighbors
import hdbscan
from sentence_transformers import SentenceTransformer
from transformers import AutoTokenizer, AutoModelForCausalLM
from typing import Dict, List
import streamlit as st

# Step 1: Load the Qwen Model and Tokenizer
tokenizer = AutoTokenizer.from_pretrained("Qwen/Qwen2.5-1.5B-Instruct", trust_remote_code=True)
model = AutoModelForCausalLM.from_pretrained("Qwen/Qwen2.5-1.5B-Instruct", trust_remote_code=True)

# Step 2: Load and Chunk Dataset
def load_and_chunk_datasets(folder_path: str) -> pd.DataFrame:
    """
    Load all JSON datasets and chunk data for embedding and clustering.
    """
    data = []
    for file in os.listdir(folder_path):
        if file.endswith('.json'):
            with open(os.path.join(folder_path, file), 'r') as f:
                content = json.load(f)
                name = content.get("name", "")
                uses = content.get("uses", "")
                side_effects = content.get("side_effects", "")
                dosage = content.get("dosage", "")

                # Chunk into smaller pieces
                chunks = [
                    {"text": f"Name: {name}. Uses: {uses}", "source": file},
                    {"text": f"Side Effects: {side_effects}", "source": file},
                    {"text": f"Dosage: {dosage}", "source": file},
                ]
                data.extend(chunks)
    return pd.DataFrame(data)

# Load the dataset
dataset_path = '/content/drive/MyDrive/Colab Notebooks/LLM Hackathon/microlabs_usa'  # Replace with the correct path
pharma_data = load_and_chunk_datasets(dataset_path)

# Step 3: Generate BERT Embeddings
bert_model = SentenceTransformer('all-MiniLM-L6-v2')
pharma_data['embedding'] = pharma_data['text'].apply(lambda x: bert_model.encode(x))

# Step 4: Clustering with HDBSCAN
embeddings = list(pharma_data['embedding'])
clusterer = hdbscan.HDBSCAN(min_cluster_size=5, metric='euclidean')
pharma_data['cluster'] = clusterer.fit_predict(embeddings)

# Step 5: Calculate Cluster Centroids
cluster_centroids = (
    pharma_data.groupby('cluster')['embedding']
    .apply(lambda x: sum(x) / len(x))
    .to_dict()
)

# Step 6: Nearest Neighbor Search
clustered_data = pharma_data[pharma_data['cluster'] != -1]  # Exclude noise
nn_models = {}
for cluster_id in clustered_data['cluster'].unique():
    cluster_subset = clustered_data[clustered_data['cluster'] == cluster_id]
    nn = NearestNeighbors(n_neighbors=5, metric='euclidean')
    nn.fit(list(cluster_subset['embedding']))
    nn_models[cluster_id] = (nn, cluster_subset)

# Assistant Class
class PharmaKnowledgeAssistant:
    def __init__(self, data: pd.DataFrame, nn_models: Dict, bert_model, cluster_centroids, tokenizer, model):
        self.data = data
        self.nn_models = nn_models
        self.bert_model = bert_model
        self.cluster_centroids = cluster_centroids
        self.tokenizer = tokenizer
        self.model = model

    def retrieve_relevant_chunks(self, query: str) -> List[Dict]:
        query_embedding = self.bert_model.encode(query)
        cluster_distances = {
            cluster_id: sum((query_embedding - centroid) ** 2)
            for cluster_id, centroid in self.cluster_centroids.items()
        }
        best_cluster = min(cluster_distances, key=cluster_distances.get)
        nn, cluster_subset = self.nn_models[best_cluster]
        distances, indices = nn.kneighbors([query_embedding])
        relevant_chunks = cluster_subset.iloc[indices[0]].to_dict(orient='records')
        return relevant_chunks

    def generate_response(self, query: str, context_chunks: List[Dict]) -> str:
        context_text = "\n".join([chunk['text'] for chunk in context_chunks])
        prompt = (
            f"Context:\n{context_text}\n\n"
            f"User Query:\n{query}\n\n"
            "Answer:"
        )
        try:
            inputs = self.tokenizer(prompt, return_tensors="pt")
            outputs = self.model.generate(inputs.input_ids, max_length=200, num_return_sequences=1)
            return self.tokenizer.decode(outputs[0], skip_special_tokens=True)
        except Exception as e:
            return f"Error generating response: {str(e)}"

    def answer_query(self, query: str) -> str:
        relevant_chunks = self.retrieve_relevant_chunks(query)
        return self.generate_response(query, relevant_chunks)

# Streamlit Interface
def main():
    st.title("Pharma Knowledge Assistant")
    st.sidebar.title("Options")
    st.sidebar.write("Upload and process pharma-related datasets.")

    # Query Input
    query = st.text_input("Enter your query:")

    if st.button("Get Answer") and query:
        assistant = PharmaKnowledgeAssistant(pharma_data, nn_models, bert_model, cluster_centroids, tokenizer, model)
        response = assistant.answer_query(query)
        st.subheader("Response:")
        st.write(response)

if __name__ == "__main__":
    main()
```
